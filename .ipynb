```python
import requests
import json
import time

# TUS NÃšMEROS
MIS_DECIMOS = ['37475', '15316', '49397', '03583', '86847', '97550', '08324', '05839', '97603']

def consultar_rtve(numero):
    # La API de RTVE es muy limpia. 
    # Devuelve: { "numero": 12345, "premio": 0, ... }
    url = f"https://api.rtve.es/loteria/numero/{numero}"
    
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0'
    }

    try:
        response = requests.get(url, headers=headers, timeout=10)
        
        # Si el nÃºmero no existe o hay error, a veces dan 404, otras veces JSON
        if response.status_code == 404:
            return 0 # No premiado o no encontrado
            
        if response.status_code != 200:
            print(f"   âš ï¸ Error Servidor: {response.status_code}")
            return None

        data = response.json()
        
        # Extraer el premio
        # La estructura suele ser directament el objeto, o dentro de 'page'
        if 'premio' in data:
            return data['premio']
        elif 'page' in data and 'items' in data['page']:
            # A veces RTVE cambia la estructura, buscamos dentro
            return data['page']['items'][0].get('premio', 0)
            
        return 0

    except Exception as e:
        print(f"   âš ï¸ Error conectando: {e}")
        return None

def main():
    print(f"--- ðŸ“» CONSULTANDO DÃ‰CIMOS EN RTVE ---\n")
    total_ganado = 0
    
    for decimo in MIS_DECIMOS:
        # RTVE a veces prefiere el nÃºmero sin ceros a la izquierda, 
        # pero probamos tal cual primero.
        premio = consultar_rtve(decimo)
        print(premio)
        if premio is not None:
            if premio > 0:
                print(f"ðŸŽ‰ {decimo} -> Â¡PREMIO! {premio}â‚¬")
                total_ganado += premio
            else:
                print(f"âŒ {decimo} -> 0â‚¬")
        else:
            print(f"âš ï¸ {decimo} -> No se pudo comprobar.")
        
        time.sleep(0.5)

    print("\n-----------------------------------")
    print(f"ðŸ’° TOTAL ACUMULADO: {total_ganado}â‚¬")
    print("(Nota: RTVE suele tardar unos minutos mÃ¡s en actualizar que la radio).")

if __name__ == "__main__":
    main()
```
