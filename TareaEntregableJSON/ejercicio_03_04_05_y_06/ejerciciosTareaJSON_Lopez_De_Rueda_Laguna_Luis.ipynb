# Ejercicio 3

El servicio web ofrecido por https://rickandmortyapi.com proporciona información sobre la serie de televisión animada “Rick and Morty”. Esta API ofrece datos detallados sobre personajes, episodios y ubicaciones dentro del universo de la serie.

A partir de esta API crea un nuevo JSON con nombre **episodios.json**, a través de un diccionario, que contenga los siguientes datos:
Identificación del episodio;

*   Identificación del episodio
*   Nombre del episodio
*   Temporada del episodio
*   Número de personajes del episodio


```python
#Cogemos todos los capitulos de todas las paginas
url = "https://rickandmortyapi.com/api/episode"
lista_todos_capitulos = []
#Hacemos un bucle while para recorrer todas las paginas
while url:
    with urllib.request.urlopen(url) as respuesta:
        cuerpo_json = respuesta.read()
    #Cargamos los apitulos y los añadimos a la lista
    cuerpo_dict = json.loads(cuerpo_json)
    lista_capitulos_actual = cuerpo_dict["results"]
    #Recorremos los capitulos de la pagina actual
    for capitulo in lista_capitulos_actual:
        #Los metemos en un diccionario.
        diccionario_capitulo = {
            "id_Capitulo": capitulo["id"],
            "nombre_Capitulo": capitulo["name"],
            "temporada_Capitulo": capitulo["episode"],
            "lista_Personajes": len(capitulo["characters"])
        }
        lista_todos_capitulos.append(diccionario_capitulo)
    
    url = cuerpo_dict["info"]["next"]
#Metemos todos los capitulos en un archivo json
with open("episodios.json", "w") as json_file:
    json.dump(lista_todos_capitulos, json_file)
```

# Ejercicio 4

El servicio web ofrecido por https://rickandmortyapi.com proporciona información sobre la serie de televisión animada “Rick and Morty”. Lee el fichero JSON, episodios.json, del ejercicio anterior, y a partir de un episodio aleatorio entre los disponibles crea un nuevo fichero JSON ,episodio_X.json, que tenga el nombre del episodio y la temporada y todos los personajes de ese episodio. La información que se tiene que obtener de cada personaje es:
*   Identificación del personaje
*   Nombre del personaje
*   Género del personaje
*   Origen del personaje
*   Localización del personaje
*   Número de episodios en los que aparece el personaje

**NOTA**: X es el episodio aleatorio que se haya obtenido

```python
with open("episodios.json", "r", encoding="utf-8") as archivo:
        datos_leidos = json.load(archivo)
capitulo_aleatorio = datos_leidos[random.randint(0, len(datos_leidos)-1)]
capitulo_aleatorio["id_Capitulo"]
url = f"https://rickandmortyapi.com/api/episode/{capitulo_aleatorio['id_Capitulo']}"
with urllib.request.urlopen(url) as respuesta:
    cuerpo_json = respuesta.read()
cuerpo_dict = json.loads(cuerpo_json)
lisa_personajes = cuerpo_dict["characters"]
lista_todos_personajes = []
for personaje_url in lisa_personajes:
    with urllib.request.urlopen(personaje_url) as respuesta:
        cuerpo_json = respuesta.read()
    cuerpo_dict = json.loads(cuerpo_json)
    diccionario_personaje = {
        "id_Personaje": cuerpo_dict["id"],
        "nombre_Personaje": cuerpo_dict["name"],
        "genero": cuerpo_dict["gender"],
        "origen_Personaje": cuerpo_dict["origin"]["name"],
        "ubicacion_Personaje": cuerpo_dict["location"]["name"],
        "numero_Apariciones": len(cuerpo_dict["episode"])
    }
    lista_todos_personajes.append(diccionario_personaje)
with open(f"episodio_{capitulo_aleatorio['id_Capitulo']}.json", "w") as json_file:
    json.dump(lista_todos_personajes, json_file)

```
