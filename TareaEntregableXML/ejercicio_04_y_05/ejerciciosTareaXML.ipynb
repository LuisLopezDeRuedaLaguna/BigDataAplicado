Utilizando el fichero **eventosDeportivos.xml**, importando la librería **ElementTree**, obten la siguiente información usando los métodos visto con esta librería:

```python
# Carga del fichero de eventos deportivos (0,25 puntos)
import xml.etree.ElementTree as ET
arbol = ET.parse('eventosDeportivos.xml')
raiz = arbol.getroot()
print("Fichero cargado correctamente.")
```

```python
# Muestra todos los títulos de los eventos (0,5 puntos)
for evento in raiz.findall('evento'):
    titulo = evento.find('titulo').text
    print(titulo)
```

```python
# Muestra el primer evento cuyo organizador es la "Real Federación Española de Fútbol".
# Debe mostrarse atributos del evento, título, fecha y lugar sin utilizar índices y sin recorrer el evento (1 punto)
for evento in raiz:
    organizador = evento.find('organizador').text
    if organizador == "Real Federación Española de Fútbol":
        print("Atributos:", evento.attrib)
        print("Título:", evento.find('titulo').text)
        print("Fecha:", evento.find('fecha').text)
        print("Lugar:", evento.find('lugar').text)
        break # Salimos del bucle tras encontrar el primero
```

```python
# Muestra los datos del último evento del XML. Ten en cuenta que el XML puede aumentar y siempre se debe mostrar el último que haya. (0,75 puntos)
# Accedemos al último elemento usando índice negativo
ultimo_evento = raiz[-1] 

print(f"Título: {ultimo_evento.find('titulo').text}")
print(f"Organizador: {ultimo_evento.find('organizador').text}")
print(f"Fecha: {ultimo_evento.find('fecha').text}")
```

```python
# Muestra en fomato texto cada uno de los datos del evento cuya asistencia es igual a 11717 (1 punto)
for evento in raiz.findall('evento'):
    if evento.get('asistentes') == '11717':
        print(f"ID: {evento.get('id')}")
        print(f"Título: {evento.find('titulo').text}")
        print(f"Lugar: {evento.find('lugar').text}")
        print(f"Ciudad: {evento.find('ciudad').text}")
```

```python
# Función que recibe el nombre de un organizador y devuelve el número de eventos de ese organizador.
# NOTA: Para probar la función no hay que introducir por teclado el texto, basta con llamar directamente a la función con el nombre el identificador (1,5 puntos)
def contar_por_organizador(nombre_organizador):
    contador = 0
    for evento in raiz.findall('evento'):
        if evento.find('organizador').text == nombre_organizador:
            contador += 1
    return contador

# Prueba
print(contar_por_organizador("Liga ACB"))
```

```python
# Función que recibe el identificador del evento y devuelve el lugar (cadena de texto) o None si no existe un evento con ese identificador.
# NOTA: Para probar la función no hay que introducir por teclado el texto, basta con llamar directamente a la función con el nombre el identificador (1,5 puntos)
def obtener_lugar_por_id(id_evento):
    for evento in raiz.findall('evento'):
        if evento.get('id') == id_evento:
            return evento.find('lugar').text
    return None

# Prueba
print(obtener_lugar_por_id("E001"))
```

```python
# Función que recibe un año y devuelve todos los títulos de los eventos de ese año. Hay que mostrar el texto de cada uno de los eventos
# NOTA: Para probar la función no hay que introducir por teclado el año, basta con llamar directamente a la función con un año (1,5 puntos)
def titulos_por_anio(anio):
    anio_str = str(anio)
    print(f"Eventos del año {anio}:")
    for evento in raiz.findall('evento'):
        if evento.get('anno') == anio_str:
            print("- " + evento.find('titulo').text)

# Prueba
titulos_por_anio(2021)
```

```python
# Añade un nuevo evento al final del todo y guarda el XML como eventosDeportivosActualizados.xml
# NOTA: No hay que introducir los nuevos datos por teclado (2 puntos)
# Crear el elemento evento con sus atributos
nuevo_evento = ET.Element('evento')
nuevo_evento.set('id', 'E999')
nuevo_evento.set('anno', '2025')
nuevo_evento.set('asistentes', '50000')

# Crear subelementos
ET.SubElement(nuevo_evento, 'organizador').text = "Comité Olímpico"
ET.SubElement(nuevo_evento, 'titulo').text = "Final 100m Lisos"
ET.SubElement(nuevo_evento, 'lugar').text = "Estadio Olímpico"
ET.SubElement(nuevo_evento, 'disciplina').text = "Atletismo"
ET.SubElement(nuevo_evento, 'ciudad').text = "Sevilla"
ET.SubElement(nuevo_evento, 'fecha').text = "2025-08-20"

# Añadir a la raíz y guardar
raiz.append(nuevo_evento)
arbol.write('eventosDeportivosActualizados.xml', encoding='utf-8')
print("Evento añadido y fichero guardado.")
```

# Ejercicio 5

Dado el fichero **sevilla.xml** con información meteorológica de la ciudad, utiliza la biblioteca **lxml** y **xpath** para obtener la siguiente información:

```python
# Carga del fichero de la información meteorológica (0,25 puntos)
from lxml import etree
doc = etree.parse('sevilla.xml')
print("Fichero sevilla.xml cargado.")
```

```python
# Muestra la latitud y longitud de la ciudad de sevilla (0,75 puntos)
latitud = doc.xpath('/tiempo/localizacion/latitud/text()')[0]
longitud = doc.xpath('/tiempo/localizacion/longitud/text()')[0]
print(f"Latitud: {latitud}, Longitud: {longitud}")
```

```python
# Muestra el texto de la temperatura, viento con su dirreción y condiciones a la fecha del pronóstico meteorológico en Sevilla (nodo condicionesAcuales) (1,25 puntos)
# Usamos el path relativo a condicionesActuales
base = '/tiempo/condicionesActuales'
temp = doc.xpath(f'{base}/temperatura/text()')[0]
viento = doc.xpath(f'{base}/viento/text()')[0]
direccion = doc.xpath(f'{base}/viento/@direccion')[0]
condiciones = doc.xpath(f'{base}/condiciones/text()')[0]

print(f"Temperatura: {temp}°C")
print(f"Viento: {viento} km/h, Dirección: {direccion}")
print(f"Condiciones: {condiciones}")
```

```python
# Muestra el número de días que me ofrece un pronóstico del tiempo. (1,25 puntos)

num_dias = doc.xpath('count(/tiempo/pronosticoDias/dia)')
print(f"Días de pronóstico: {int(num_dias)}")
```

```python
# Obtén las fechas dentro del pronóstico en las que la velocidad del viento es inferior a 10. Muestra cada una de las fechas por separado, no es una lista (1,25 puntos)
fechas = doc.xpath('/tiempo/pronosticoDias/dia[viento < 10]/@fecha')

for fecha in fechas:
    print(fecha)
```

```python
# A partir de los datos registrados, etiqueta "datosRegistrados", obtén la presión media con dos decimales (1,25 puntos)

presiones = doc.xpath('/tiempo/datosRegistrados/dato/presion/text()')
lista_presiones = [float(p) for p in presiones]
media = sum(lista_presiones) / len(lista_presiones)

print(f"Presión media: {media:.2f}")
```

```python
# A partir de los datos registrados, etiqueta "datosRegistrados", obtén la fecha o fechas de máxima humedad.
# NOTA: La función max puede que no esté soportada dentro de XPATH (2 puntos)
humedades = doc.xpath('/tiempo/datosRegistrados/dato/humedad/text()')
humedades_int = [int(h) for h in humedades]
max_humedad = max(humedades_int)
fechas_max_humedad = doc.xpath(f'/tiempo/datosRegistrados/dato[humedad="{max_humedad}"]/@fecha')

print(f"Humedad máxima ({max_humedad}%) encontrada en: {fechas_max_humedad}")
```

```python
# Crea una función que te devuelva un diccionario con la temperatura y la sensación térmica de las horas que me ofrece el pronóstico.
# Utiliza como clave del diccionario la hora y como valor una tupla formada por la temperatura y la sensación térmica. (2 puntos)
def obtener_dict_temperaturas():
    datos = {}
    nodos_hora = doc.xpath('/tiempo/pronosticoHoras/hora')
    
    for nodo in nodos_hora:
        hora_id = nodo.get('id')
        temp = nodo.findtext('temperatura')
        sensacion = nodo.findtext('sensacionTermica')
        datos[hora_id] = (temp, sensacion)
        
    return datos
diccionario = obtener_dict_temperaturas()
print(diccionario)
```
